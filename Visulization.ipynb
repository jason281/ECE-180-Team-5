```python
import csv
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_file
```

```python
people_csv = open('dataset/Cleanedup_people.csv','r')
title_csv = open('dataset/Cleanedup_titles.csv','r')
people_reader = csv.reader(people_csv,delimiter='\t')
title_reader = csv.reader(title_csv,delimiter='\t')
```

```python
idx = 0
row = next(title_reader, None)
row = next(title_reader, None)
genre_rate = {}
director_rate = {}
score = []
numVote = []
while row:
    genres = row[9].split(',')
    dirs = row[13].split(',')
    for g in genres:
        if g == '\\n' or g=='NaN':
            continue
        if g not in genre_rate:
            genre_rate[g] = []
        genre_rate[g].append(float(row[10]))
    for d in dirs:
        if d == '\\n' or d=='NaN':
            continue
        if d not in director_rate:
            director_rate[d] = []
        director_rate[d].append(float(row[10]))
    if row[10]!='NaN' and row[10]!='\\n' and row[11]!='NaN' and row[11]!='\\n':
        score.append(float(row[10]))
        numVote.append(float(row[11]))
    idx += 1
    row = next(title_reader, None)
```

```python
genre_weight = [len(genre_rate[i]) for i in genre_rate]
plt.figure()
plt.pie(genre_weight, labels=genre_rate.keys())
plt.savefig('genre_pie.png')
plt.show()
```

```python
plt.figure()
plt.boxplot(genre_rate.values(), labels=genre_rate.keys())
plt.savefig('genre_box.png')
plt.show()
```

```python
plt.figure()
plt.scatter(numVote, score)
plt.savefig('numVote_vs_rate.png')
plt.show()
```
