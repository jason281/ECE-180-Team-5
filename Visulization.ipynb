```python
import csv
import matplotlib.pyplot as plt
```

```python
people_csv = open('dataset/Cleanedup_people.csv','r')
title_csv = open('dataset/Cleanedup_titles.csv','r')
people_reader = csv.reader(people_csv,delimiter='\t')
title_reader = csv.reader(title_csv,delimiter='\t')
```

```python
idx = 0
row = next(title_reader, None)
row = next(title_reader, None)
genre_rate = {}
director_rate = {}
while row:
    genres = row[9].split(',')
    dirs = row[13].split(',')
    for g in genres:
        if g not in genre_rate:
            genre_rate[g] = []
        genre_rate[g].append(float(row[10]))
    for d in dirs:
        if d not in director_rate:
            director_rate[d] = []
        director_rate[d].append(float(row[10]))
    idx += 1
    row = next(title_reader, None)
```

```python
plt.figure()
plt.boxplot(genre_rate.values())
```

```python
plt.figure()
plt.boxplot(director_rate.values())
```

```python
people_csv.close()
title_csv.close()
```
