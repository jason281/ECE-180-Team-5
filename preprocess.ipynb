```python
import pandas as pd
import random
import numpy as np
```

```python
def parseData(filename):
    return pd.read_csv('../dataset/%s'%(filename), compression='gzip',delimiter='\t')
```

```python
def createHash(data,key,value):
    return dict(zip(data[key],data[value]))
```

# People Behavior & Name Hashtable 

```python
data_people = parseData('name.basics.tsv.gz');
```

```python
name_dict = createHash(data_people,'nconst','primaryName')
```

# Video Behavior & Title Hashtable

```python
data_video = parseData('title.basics.tsv.gz');
```

```python
title_dict = createHash(data_video,'tconst','primaryTitle')
```

# Video Ratings #

```python
data_ratings = parseData('title.ratings.tsv.gz');
```

# Cast of Each Video #

```python
data_cast = parseData('title.principals.tsv.gz');
```

# Director & Writor of Each Video #

```python
data_crew = parseData('title.crew.tsv.gz');
```

# Merge datasets of Video #

Ordered by tconst

```python
data_merge = data_video
data_add = [data_ratings,data_cast,data_crew]
for i in range(len(data_add)):
    data_merge = pd.merge(data_merge,data_add[i],how = 'left',on='tconst')
```

```python
data_merge
```

# Re-hashing title-name const #

Rehash video title in data_people

```python
def rehash(hashtable,column):
    '''
    para: hashtable, column
    type: dict, pd.DataFrame column
    
    ouput: rehashlist
    type: list
    
    p.s. output is in lower-letter form
    '''
    rehashlist = []
    
    for d in column:
        d = str(d)
        tmp = []
        trunc = d.split(',')
        for r in trunc:
            if r=='\\N' or r == 'nan':
                tmp.append('\N')
            
            else: 
                try:
                    tmp.append(hashtable[r].lower())
                except:
                    tmp.append('ERROR:OUTSIDE DICTIONARY') 
             
        rehashlist.append(','.join(tmp))
    
    return rehashlist
```

```python
rehash_in_people = rehash(title_dict,data_people['knownForTitles'])
```

```python
data_people['knownForTitles'] = rehash_in_people
```

```python
data_people
```

Rehash people name in title data

```python
rehash_in_cast = rehash(name_dict,data_merge['principalCast'])
rehash_in_director = rehash(name_dict,data_merge['directors'])
rehash_in_writer = rehash(name_dict,data_merge['writers'])
```

```python
data_merge['principalCast'] = rehash_in_cast
data_merge['directors'] = rehash_in_director
data_merge['writers'] = rehash_in_writer
```
